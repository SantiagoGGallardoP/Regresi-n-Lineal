## FILTRO 1 & 2. “host_acceptance_rate vs host_response_rate”

```python
# Eliminar los símbolos de porcentaje
df_mex['host_acceptance_rate'] = df_mex['host_acceptance_rate'].str.replace('%', '')
df_mex['host_acceptance_rate'] = df_mex['host_acceptance_rate'].str.replace('omitido', "94.44066297313655")
# Convertir la columna a tipo float
df_mex['host_acceptance_rate'] = df_mex['host_acceptance_rate'].astype(float)
promedio = df_mex.loc[df_mex['host_acceptance_rate'] >= 1, 'host_acceptance_rate'].mean()
promedio
```

```python
df_mex["host_response_rate"] = df_mex["host_response_rate"].str.replace('%', '')
df_mex["host_response_rate"] = df_mex["host_response_rate"].str.replace('omitido', "97.08")
# Convertir la columna a tipo float
df_mex["host_response_rate"] = df_mex["host_response_rate"].astype(float)
promedio = df_mex.loc[df_mex["host_response_rate"] >= 1, "host_response_rate"].mean()
promedio
```

```python
filtro = df_mex["room_type"] == "Entire home/apt"
entire_home_mex = df_mex[filtro]
entire_home_mex
```

```python
filtro2 = df_mex["room_type"] == "Private room"
private_room_mex = df_mex[filtro2]
private_room_mex
```

```python
#Imprimimos el scatter plot entre la variable dependiente (total) e independiente (alchool) 
#para observar el comportamiento en su dispersión 
from turtle import color

sns.scatterplot(y="host_acceptance_rate", x="host_response_rate", color="blue", data=entire_home_mex)
#sns.scatterplot(X='speeding', y='total', color="red", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
#Vars_Indep= df[['alcohol', 'speeding']]
Vars_Indep= entire_home_mex[["host_response_rate"]]
Var_Dep= entire_home_mex["host_acceptance_rate"]
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
model.score(Vars_Indep,Var_Dep)
```

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
